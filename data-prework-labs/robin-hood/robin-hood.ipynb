## Description:
In the 2-dimensional space, a point can be defined by a pair of values that correspond to the horizontal coordinate (x) and the vertical coordinate (y). The space can be divided into 4 zones (quadrants): Q1, Q2, Q3, Q4. Whose single point of union is the point (0, 0).

If a point is in Q1 both its x coordinate and the y are positive. I leave a link to wikipedia to familiarize yourself with these quadrants.

## Tasks
1. Robin Hood is famous for hitting an arrow with another arrow. Did you get it?
2. Calculate how many arrows have fallen in each quadrant.
3. Find the point closest to the center. Calculate its distance to the center. 
4. If the target has a radius of 9, calculate the number of arrows that must be picked up in the forest.

````python
# Variables

points = [(4, 5), (-0, 2), (4, 7), (1, -3), (3, -2), (4, 5),
          (3, 2), (5, 7), (-5, 7), (2, 2), (-4, 5), (0, -2),
          (-4, 7), (-1, 3), (-3, 2), (-4, -5), (-3, 2),
          (5, 7), (5, 7), (2, 2), (9, 9), (-8, -9)]

# 1. Robin Hood is famous for hitting an arrow with another arrow. Did you get it?

print(len(points)) #checks the length of the list
set_points = (set(points)) # transforms list into a set
print(len(set_points)) #check the lenght of the set; it's smaller because set removes duplicates

if len(points) != len(set_points):  
    print("The length of the list is " + str(len(points)) + ". The length of the set is " + str(len(set_points)) + ".")

# 2. Calculate how many arrows have fallen in each quadrant.

points_q1 = 0
points_q2 = 0
points_q3 = 0
points_q4 = 0


for point in points:
    x = point[0]
    y = point[1]
    
    if x > 0 and y > 0: 
        points_q1 += 1
    
    elif x < 0 and y > 0:
        points_q2 += 1
    
    elif x < 0 and y < 0:
        points_q3 += 1
    else:
        points_q4 += 1
    
print("There were " + str(points_q1) +" shots on Q1.")
print("There were " + str(points_q2) +" shots on Q2.")
print("There were " + str(points_q3) +" shots on Q3.")
print("There were " + str(points_q4) +" shots on Q4.")

# 3. Find the point closest to the center. Calculate its distance to the center
# Defining a function that calculates the distance to the center can help.

distances_center = []

for point in points: 
    x = point[0]
    y = point[1]
    
    way_to_center = (abs(x) ** 2 + abs(y) ** 2) ** 0.5
     
    distances_center.append(way_to_center)
    
print(min(distances_center))

min_center = distances_center.index(min(distances_center))

print(min_center)
print ("The closest point to the center is "+ str(points[min_center]) + ".")
    
# 4. If the target has a radius of 9, calculate the number of arrows that 
# must be picked up in the forest.
````
