#### Display the top 10 languages by number of native speakers stored in a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
# your code here
html = requests.get(url).content

native_soup = BeautifulSoup(html, "lxml") 

#there's more than one table on the page, so I'm selecting the first one

table = native_soup.find_all('table', attrs = {'class':'wikitable sortable'})[0]
table
````

````python
#cleaning the header
header = [content.text for content in table.find_all('th')]
header = [item.replace('\n', ' ').replace('\xa0', '').split('  ') for item in header]
flat_header = [item for sublist in header for item in sublist]
flat_header = [item for item in flat_header if item]

print(flat_header)
````

````python
rows = table.find_all('tr')

#tidying stuff up; it's not super tidy, though

rows = [row.text.replace('\n', ' ').replace('\xa0', '').split('  ') for row in rows[1:]]

wiki_df = pd.DataFrame(rows)

#creating dict for header renaming

keys = wiki_df.columns
values = flat_header
header_dict = dict(zip(keys, values))
wiki_df.rename(columns = header_dict, inplace=True)
wiki_df
````

#### Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
html = requests.get(url).content

imdb_soup = BeautifulSoup(html, "html.parser")

cine_table = imdb_soup.find('table', attrs = {'data-caller-name': 'chart-top250movie'})
cine_rows = cine_table.find_all('tr')

cine_table
````

````python
# getting the rest of the stuff

titles = [t.find_all('img') for t in cine_table if t.find('img')]
titles
````

````python
full_cast = [title.a.get('title').replace(' (dir.)', '').split(', ') for title in cine_table if (title.a and title.a.get('title'))]
director = [x[0] for x in full_cast]
cast = [x[1:] for x in full_cast]

years = [int(title.text[1:-1]) for title in imdb_soup.find_all('span', {'class':'secondaryInfo'})]

cine_chart = pd.DataFrame({'Title' : title, 'Year' : years, 'Director' : director, 'Cast' : cast})

#cine_chart = pd.DataFrame({'Director' : director, 'Cast' : cast})
cine_chart
````
